```python
# import statements
import sys
import numpy as np
import matplotlib.pyplot as plt
import scipy
import os
from os import listdir
from skimage import io
from PIL import Image
from scipy.misc import imresize
from scipy import ndimage
```

```python
path = "images"
diff = ["\\cats\\","\\dogs\\"]
values = [0,1,2]
X,Y = [],[]

def get_img(data):  #return the image in np array format
    img_size = 64
    img = (Image.open(data))
    img = img.resize((img_size, img_size))
    if data == "1.jpeg":
        img.show()
    img = np.asarray(img)
    return img
```

```python
def get_data():
    
    try:                              #loads the stored data
        X = np.load('Data\X.npy')
        Y = np.load('Data\Y.npy')
    except:                            #reads the images and stores them in X and Y
        count = 0
        #for data in listdir("images\lego"):
         #   print("images\lego\\"+data)
        len_data = 0
        for data_v in listdir(path):
            len_data += len(listdir(path+"//"+data_v))
        print(len_data)
        X = np.zeros((len_data,64,64,3),dtype='float64')
        Y = np.zeros((len_data,1))
        print(X.shape)
        print(Y.shape)
        j_count = 0
        for j in diff:
            for i in listdir(path+j):
                temp_img = get_img(path+j+i)
                #print(temp_img)
                #print(temp_img.shape)
                print("Loading img "+ str(count))
                X[count]= temp_img
                Y[count]= values[j_count]
                count += 1
            j_count += 1
        X /= 255
        if not os.path.exists('Data'):
            os.makedirs('Data')
        np.save('Data\X.npy',X)
        np.save('Data\Y.npy',Y)
        print(Y)
    return X,Y
```

```python

def flatten(X_train, X_test, Y_train, Y_test):
    X_train = X_train.reshape(X_train.shape[0],-1).T
    X_test = X_test.reshape(X_test.shape[0],-1).T
    Y_train = Y_train.T
    Y_test = Y_test.T
    return X_train, X_test, Y_train, Y_test
```

```python

def sigmoid(z):      #sigmoid function    
    return 1/(1+np.exp(-z))
```

```python

def initialize_with_zeros(dim):
    w = np.zeros((dim,1))     #creates weight matrix of required size initialized to 0
    b = 0                     #creates a bais component initialized to 0 
    #print("shape of w",w.shape)
    #print("b ",b)
    
    return w,b
```

```python


def propagate(w, b, X, Y):

    m = X.shape[1]
    #print("m ",m)
    A = sigmoid(np.dot(w.T,X) + b)    
    #print("a",A.shape)# compute activation
    #print("y",Y.shape)
    #print("w , X",w.shape,X.shape)
    cost = (-1/m)*(np.sum(Y*np.log(A)+(1-Y)*np.log(1-A)))          # compute cost
    #print(cost)
    dw = (1/m)*(np.dot(X,(A-Y).T))
    db = (1/m)*(np.sum(A-Y))
    cost = np.squeeze(cost)
    
    assert(dw.shape == w.shape)
    assert(db.dtype == float)
    cost = np.squeeze(cost)
    assert(cost.shape == ())
    grads = {"dw": dw,
             "db": db}
    return grads, cost
```

```python

def optimize(w, b, X, Y, num_iterations, learning_rate):
    
    costs = []
    
    for i in range(num_iterations):
        grads, cost = propagate(w, b, X, Y)
  
        dw = grads["dw"]
        db = grads["db"]
        w = w - learning_rate*(dw)
        b = b - learning_rate*(db)

        if i % 100 == 0:
            costs.append(cost)
        if  i % 100 == 0:
            print ("Cost after iteration %i: %f" %(i, cost))
    #print(costs)
    params = {"w": w,
              "b": b}
    
    grads = {"dw": dw,
             "db": db}
    
    return params, grads, costs
```

```python

def predict(w, b, X):
    m = X.shape[1]
    Y_prediction = np.zeros((1,m))
    w = w.reshape(X.shape[0], 1)
    A = sigmoid(np.dot(w.T,X)+b)
    for i in range(A.shape[1]):
        if A[0,i] <= 0.5:
            Y_prediction[0,i] = 0
        else:
            Y_prediction[0,i] = 1
    return Y_prediction
    
```

```python

def build_model(X_train, X_test, Y_train, Y_test,learning_rate = 0.5,num_iterations = 2000):
    
    w, b = initialize_with_zeros(X_train.shape[0])
        
    parameters, grads, costs = optimize(w, b, X_train, Y_train, num_iterations, learning_rate)
    
    w = parameters["w"]
    b = parameters["b"]    
    
    Y_prediction_test = predict(w, b, X_test)
    Y_prediction_train = predict(w, b, X_train)
    
    print("train accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_test - Y_test)) * 100))

    
    d = {"costs": costs,
         "Y_prediction_test": Y_prediction_test, 
         "Y_prediction_train" : Y_prediction_train, 
         "w" : w, 
         "b" : b,
         "learning_rate" : learning_rate,
         "num_iterations": num_iterations}
    
    return d
```

```python


def main():              
    X,Y = get_data() #load the datasets
    #print(X.shape)
    #print(Y.shape)
    from sklearn.model_selection import train_test_split
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.05, random_state=42) # get training and testing data
    #print(X_train.shape)
    #print(X_test.shape)
    #print(Y_train.shape)
    X_train, X_test, Y_train, Y_test = flatten(X_train, X_test, Y_train, Y_test)
    d = build_model(X_train, X_test, Y_train, Y_test,learning_rate = 0.005,num_iterations = 2000)
    my_image = "2.jpg"   # change this to the name of your image file 

    fname = "check/" + my_image
    image = np.array(ndimage.imread(fname, flatten=False))
    my_image = scipy.misc.imresize(image, size=(64,64)).reshape((1, 64*64*3)).T
    my_predicted_image = predict(d["w"], d["b"], my_image)

    plt.imshow(image)
    print("y = " + str(np.squeeze(my_predicted_image)) + ", your algorithm predicts a")
    for i in my_predicted_image:
        if i == 0:
            print("cat")
        else:
            print("dog")

    
    
if __name__ == "__main__":
    main()
```
